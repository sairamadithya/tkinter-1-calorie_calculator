```python
def exercise_calculator(walk,run,push,squat,cycle,cross):
    c1= (5*walk)+(9*run)+(7*push)+(8*squat)+(7*cycle)+(16*cross)
    return c1   
print("enter the minutes or enter 0 if not done")
print("enter the minutes of walking")
walk= int(input())
print(" enter the minutes of running")
run=int(input())
print("enter the minutes of pushup")
push= int(input())
print("enter the minutes of squats")
squat= int(input())
print("enter the minutes of cycling")
cycle= int(input())
print("enter the minutes of crossfit")
cross= int(input())
calories= exercise_calculator(walk,run,push,squat,cycle,cross)
print("the total number of calories burnt is",calories)    
```
